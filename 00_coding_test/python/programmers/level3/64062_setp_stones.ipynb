```python
def is_over(mid, stones, k):
    count = 0
    for stone in stones:
        if stone - mid <= 0:
            count += 1
            if count >= k:  # 연속된 장애물이 k 이상이면 True
                return True
        else:
            count = 0
    return False  # 연속된 장애물이 k 미만

def solution(stones, k):
    left = 1  # 최소 1명부터 시작
    right = max(stones)
    answer = 0

    while left <= right:
        mid = (left + right) // 2
        if is_over(mid, stones, k):  # mid명이 건널 수 없는 경우
            answer = mid  # 가능한 값을 저장
            right = mid - 1
        else:  # mid명이 건널 수 있는 경우
            left = mid + 1

    return answer
```
